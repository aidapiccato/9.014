# Problem 1: The Hill Function as a Prototype for the Sigmoid Function
In this exercise you will pratice basic aspects of Python by studying the Hill function. The HIll function was formulated to capture the sigmoidal nature of [oxygen binding to hemoglobin]((https://en.wikipedia.org/wiki/Hill_equation_(biochemistry))). Nonetheless, this function is generalizable to other scenarios. For example, it can be used to model [neural computations in the visual cortex](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3273486/).

The basic form of the Hill function is: 

$$
f(x) = \frac{x^n}{K^n + x^n}
$$

where $x$ is the only independent variable that is assumed to be greater or equal than 0. $K$ and $n$ are two positive parameters that control the shape of the sigmoid curve. The interpretability of these parametrs is dependent on what we are trying to model.

In the following steps we will build an intuition of what these parameters do. This will also provide an opportunity to practice basic Python operations and plotting.
