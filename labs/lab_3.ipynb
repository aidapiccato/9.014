# Multivariate Normal Distributions
## Introduction
The multivariate normal distribution is a generalization of the univariate normal distribution to two or more variables. It has two parameters, a mean vector μ and a covariance matrix Σ, that are analogous to the mean and variance parameters of a univariate normal distribution. The diagonal elements of Σ contain the variances for each variable, and the off-diagonal elements of Σ contain the covariances between variables.
The probability density function (PDF) of the $d$-dimensional multivariate normal distribution is:

$$
y = f(x, \mu, \Sigma) = \frac{1}{|\Sigma|(2\pi)^d}\exp{\bigg(-\frac{1}{2}(x - \Sigma)\Sigma^{-1}(x - \mu)}\bigg)
$$

where x and μ are 1-by-d vectors and Σ is a d-by-d symmetric, positive definite matrix. 

The Python command `np.random.multivariate_normal(mean, cov, size)` (documentation [here](https://numpy.org/doc/stable/reference/random/generated/numpy.random.multivariate_normal.html)) returns a matrix of shape `size`, where each entry in the matrix is an $N$-dimensional value drawn from the distribution defined by the mean and covariance vector `mean` and `cov`. 
- This function requires the matrix `cov` to be symmetric. If `cov`has only minor asymmetry, you an use `(cov + cov.T')/2` instead to resolve the asymmetry

The function `scipy.stats.multivariate_normal` (documentation [here](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.multivariate_normal.html)) also defines a multivariate normal, and is useful for finding the PDF.
